Memanggil library yang digunakan

Import data csv/txt/xls dari file directory menggunakan library pandas

```
data = pd.read_csv('/content/test.csv')
```

Menampilkan 5 baris teratas bertujuan: agar mendapat penggambaran data apa saja yang kita miliki

Melihat apakah ada nilai berupa NA/NaN dalam data

```
data.info()
```

Dilihat dari RangeIndex seharusnya data entries/baris adalah 1459 non-null (not NA/NaN) tersebar pada 80 kolom. Tetapi pada data.info() terlihat bahwa beberapa kolom tidak memunculkan jumlah entries(jumlah baris) yang sama, sehingga hal ini menunjukkan bahwa pada kolom tersebut ada nilai NA/NaN 

Jika ingin menghapus semua kolom yang mengandung nilai NA/NaN

```
data_f = data.dropna(axis=1)
#melihat list data_f
data_f
```

Jika ingin menghapus semua baris yang mengandung nilai NA/NaN

```
data_p = data.dropna(axis=0)
#melihat list data_p
data_p
```

Jika ingin menghapus data NaN pada suatu baris/kolom tertentu

```
data_s = data.dropna(how='any')
#melihat list data_s
data_s
```

Jika ingin menghapus semua baris/kolom yang mengandung NaN

```
data_d = data.dropna(how='all')
#melihat list data_d
data_d
```

Jika ingin mengisi data NaN dengan nilai rata-rata (Apabila list dari data berupa kategori, maka perlu dilakukan konversi variabel kategory (misal a,b,c,d) ke variable indikator (0, 1, 2, dst), namun apabila nilai list pada data keseluruhan adalah angka bukan kategori, maka konversi tidak perlu dilakukan)

```
#Konversi variabel kategori ke variabel indikator
dataset = pd.get_dummies(data)
# Mengisi data NaN dengan nilai rata-rata:
dataset = dataset.fillna(data.mean())
#melihat list dataset:
dataset
```

Melihat apakah ada nilai berupa NA/NaN dalam data setelah diisi dengan nilai rerata

```
dataset.info()
```

Menampilkan jumlah nilai NA/NaN dalam data per kolom

```
print(dataset.isnull().sum())
```

Jika ingin mengganti nilai NaN dengan data hasil forward propagation, yaitu mengambil acuan dari data urutan diatasnya

```
#Konversi variabel kategori ke variabel indikator
dataset_2 = pd.get_dummies(data)
# Mengisi data NaN dengan mengacu dari data urutan diatasnya:
dataset_2 = dataset_2.fillna(method='ffill')
#melihat list dataset:
dataset_2
```

Jika ingin mengganti nilai NaN dengan data hasil back propagation, yaitu mengambil acuan dari data urutan dibawahnya

```
#Konversi variabel kategori ke variabel indikator
dataset_3 = pd.get_dummies(data)
# Mengisi data NaN dengan mengacu dari data urutan dibawahnya:
dataset_3 = dataset_3.fillna(method='bfill')
#melihat list dataset:
dataset_3
```

Jika ingin mengganti nilai NaN dengan angka 0

```
#Konversi variabel kategori ke variabel indikator
dataset_4 = pd.get_dummies(data)
# Mengisi data NaN dengan mengacu dari data urutan dibawahnya:
dataset_4 = dataset_4.fillna(0)
#melihat list dataset:
dataset_4
```

Jika ingin menggunakan interpolasi linear (dengan menggunakan konversi kategori ke indikator)

```
#Konversi variabel kategori ke variabel indikator
dataset_5 = pd.get_dummies(data)
# Metode interpolasi linear
dataset_5 = dataset_5.interpolate(method='linear')
#melihat list dataset:
dataset_5
```

Jika ingin menggunakan interpolasi linear (tanpa menggunakan konversi kategori ke indikator) = disini nilai NaN masih ada, karena interpolasi bukan mengganti/mengisi data NaN melainkan mengganti data tertentu sesuai hasil interpolasi linear.

```
dataset_6 = data.interpolate(method='linear')
#melihat list dataset:
dataset_6.info()
```

Solusi masalah diatas yaitu dengan menggunakan nilai threshold pada interpolasi metodi pad untuk membatasi maksimum jumlah NaN dikehendaki.



```
dataset_7 = data.interpolate(method='pad', limit=7)
#melihat list dataset:
dataset_7.info()
```

NaN sudah berkurang namun masih terdapat NaN pada 3 kolom, dimana jumlah baris yang non Nan bukan 1459:
1.   LotFrontage    1232 non-null   float64 
2.   MasVnrArea     1455 non-null   float64
3.   MSZoning       1455 non-null   object







Karena kedua kolom tersebut berisi nilai float atau angka dan satu kolom berisi object, sehingga solusi diatas yaitu kembali ke solusi sebelumnya yaitu dengan mengisi nilai angka dengan rerata, forward propagation, backward propagation, nol, dan object dengan metode interpolasi.

```
dataset_7['LotFrontage'] = dataset_7['LotFrontage'].fillna(dataset_7.mean)
dataset_7['MasVnrArea'] = dataset_7['MasVnrArea'].fillna(dataset_7.mean)
dataset_7['MSZoning'] = dataset_7['MSZoning'].interpolate(method='pad', limit = 3)
dataset_7.info()
```

```
print(dataset_7.isnull().sum())
```

Sehingga kita dapatkan data tanpa NaN
